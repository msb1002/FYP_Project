# Dummy Generation

```python
import random
import pandas as pd
```

```python

def generator():
  age = random.randint(0,100)
  portion = random.randint(0,100)
  item = random.choice(["0-10k", "10-20k", "20-40k","40-80k",">80k"])
  risk = random.choice(['0-5%', '5-10%', '10-20%','20-30%','over 30%'])
  why = random.choice(['1. Stock as a part of portfolio', '2. Only source of profit', '3. As a part of experience :)'])
  term = random.choice(['Very short (1 month)','Short Term (within 3 months)', 'Regular (6month~1 year)', 'Quite Long (1~3 year)','Long (More than 3 years)'])
  bearing = random.randint(1,10)
  t = random.randint(1,4)
  types = random.sample(['Leverage', 'Individual Stocks', 'ETF','Fund'],t)
  profit_exp = random.choice(["0-10%",	"10-20%",	"20-40%",	"40-80%",	">80%"])
  conf = random.randint(1,10)
  res_1 = random.choice(['1. Sell everything and give up', '2. Sell majority of stock and take some time before re-enter', '3. Take out 50%, and wait before the rebound', 'Volatility is inevitable, so only sell the weak assets', '5. Remain calm and keep the portfolio as is'])
  res_2 = random.choice(['1. Sell everything and gain profit', '2. Sell majority of stock and take some time before re-enter', '3. Take out 50%, and wait for further growth', 'Volatility is inevitable, so only sell the least increased assets', '5. Keep the portfolio as is'])
  tf1 = random.choice(['True', 'False', "Don't know"])
  tf2 = random.choice(['True', 'False', "Don't know"])
  tf3 = random.choice(['True', 'False', "Don't know"])
  tf4 = random.choice(['True', 'False', "Don't know"])

  stock_type = ['Blockchain','Airline','Manufacturing','ESG','Agricultural','Traveling','Semiconductors',
  'Artificial Intelligence','Cloud computing','Biotechnology',' Social Media',' Entertainment',
  'Retail','Franchise','Real estate','Telecommunication','Energy & Resources','Commodity','Luxury goods']
  st = random.randint(1,15)
  stock_type = random.sample(stock_type,st)

  return age,portion,item,risk,why,term,bearing,types,profit_exp,conf,res_1,res_2,tf1,tf2,tf3,tf4,stock_type
```

```python
column_list = ['age','portion','item','risk','why','term','bearing','types','profit_exp','conf','res_1','res_2','tf1','tf2','tf3','tf4','stock_type']
dummy_df = pd.DataFrame(columns=column_list)
```

```python
for i in range(1,1001):
  inputs = generator()
  dummy_df.loc[i] = inputs
```
